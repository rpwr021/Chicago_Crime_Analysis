```python
crime_reports = {}
for element in all_links:
    crime_reports[element.a.get_text()] = {}
    crime_reports[element.a.get_text()]["link"] = element.a["href"]
    vname = element.a.contents[2].strip().replace(" ", "_")
    crime_reports[element.a.get_text()]["victim_name"] = vname
    r0 = urllib.request.urlopen(element.a["href"]).read()
    soup0 = BeautifulSoup(r0)
    report = soup0.find_all('p',)
    s = ''
    for i in range(0,len(report)):
       s +=  report[i].get_text() 
    crime_reports[element.a.get_text()]["news_text"] = s

    
    
```

```python
import pandas as pd
```

```python
cdata = pd.DataFrame.from_dict(crime_reports)
```
